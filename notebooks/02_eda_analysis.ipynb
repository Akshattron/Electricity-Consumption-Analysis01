```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
plt.style.use('seaborn-v0_8')
plt.rcParams['figure.dpi'] = 100
print(" All libraries imported successfully!")
print(" Ready for Exploratory Data Analysis")
```

```python
df = pd.read_csv(
    "cleaned_household_power.csv",
    index_col="DateTime",
    parse_dates=True
)

print(" Cleaned data loaded successfully!")
print(f" Total records: {len(df):,}")
print(f" Date range: {df.index.min()} to {df.index.max()}")
print(f" Total columns: {len(df.columns)}")
```

```python
print("üëÄ First 10 rows of the dataset:")
print("="*80)
df.head(10)
```

```python
print("üìã Dataset Information:")
print("="*80)
df.info()
```

```python
print("üìà STATISTICAL SUMMARY")
print("="*80)
print("\nThis shows mean, min, max, and standard deviation for all features")
print("-"*80)

summary = df.describe()
print(summary)

print("\nüîç KEY INSIGHTS:")
print("="*80)
print(f" Average Global Active Power: {df['Global_active_power'].mean():.3f} kW")
print(f" Maximum Power Consumption: {df['Global_active_power'].max():.3f} kW")
print(f" Minimum Power Consumption: {df['Global_active_power'].min():.3f} kW")
print(f" Standard Deviation: {df['Global_active_power'].std():.3f} kW")
```

```python
print(" Missing Values Check:")
print("="*80)
missing = df.isnull().sum()
print(missing)

if missing.sum() == 0:
    print("\n No missing values found! Data is clean.")
else:
    print(f"\n Total missing values: {missing.sum()}")
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(df['Global_active_power'], bins=50, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Distribution of Global Active Power', fontsize=14, fontweight='bold')
plt.xlabel('Power (kW)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', alpha=0.3)

plt.subplot(1, 2, 2)
plt.hist(df['Global_active_power'], bins=50, color='coral', edgecolor='black', alpha=0.7, density=True)
plt.title('Distribution with Density', fontsize=14, fontweight='bold')
plt.xlabel('Power (kW)', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('distribution_global_power.png', dpi=300, bbox_inches='tight')
plt.show()

print(" Graph 1 saved: distribution_global_power.png")
print(" This shows the most common electricity usage levels")
```

```python
daily_usage = df['Global_active_power'].resample('D').mean()

plt.figure(figsize=(14, 6))
plt.plot(daily_usage, linewidth=1.5, color='darkblue', alpha=0.8)
plt.title('Daily Average Electricity Consumption', fontsize=16, fontweight='bold')
plt.xlabel('Date', fontsize=12)
plt.ylabel('Power (kW)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('daily_consumption.png', dpi=300, bbox_inches='tight')
plt.show()

print(" Graph 2 saved: daily_consumption.png")
print(" This shows day-to-day electricity usage patterns")
print(f" Average daily consumption: {daily_usage.mean():.3f} kW")
```

```python
monthly_usage = df['Global_active_power'].resample('M').mean()

plt.figure(figsize=(14, 6))
plt.plot(monthly_usage, marker='o', linewidth=2, markersize=6, color='darkgreen')
plt.title('Monthly Average Electricity Consumption', fontsize=16, fontweight='bold')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Power (kW)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('monthly_consumption.png', dpi=300, bbox_inches='tight')
plt.show()

print(" Graph 3 saved: monthly_consumption.png")
print(" This reveals seasonal patterns (summer vs winter)")
print(f" Highest monthly average: {monthly_usage.max():.3f} kW")
print(f" Lowest monthly average: {monthly_usage.min():.3f} kW")
```

```python
sub_daily = df[['Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']].resample('D').mean()

plt.figure(figsize=(14, 6))
plt.plot(sub_daily['Sub_metering_1'], label='Kitchen', linewidth=1.5, alpha=0.8)
plt.plot(sub_daily['Sub_metering_2'], label='Laundry', linewidth=1.5, alpha=0.8)
plt.plot(sub_daily['Sub_metering_3'], label='Water Heater & AC', linewidth=1.5, alpha=0.8)

plt.title('Daily Sub-Metering Electricity Usage (Household Areas)', fontsize=16, fontweight='bold')
plt.xlabel('Date', fontsize=12)
plt.ylabel('Energy (Watt-hour)', fontsize=12)
plt.legend(loc='best', fontsize=11)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('submetering_comparison.png', dpi=300, bbox_inches='tight')
plt.show()

print(" Graph 4 saved: submetering_comparison.png")
print(" This compares electricity usage across different household areas")
print("\n Average consumption by area:")
print(f"   üî∏ Kitchen (Sub_metering_1): {df['Sub_metering_1'].mean():.2f} Wh")
print(f"   üî∏ Laundry (Sub_metering_2): {df['Sub_metering_2'].mean():.2f} Wh")
print(f"   üî∏ Water Heater & AC (Sub_metering_3): {df['Sub_metering_3'].mean():.2f} Wh")
```

```python
plt.figure(figsize=(10, 6))
plt.boxplot(df['Global_active_power'], vert=True, patch_artist=True,
            boxprops=dict(facecolor='lightblue', color='darkblue'),
            whiskerprops=dict(color='darkblue'),
            capprops=dict(color='darkblue'),
            medianprops=dict(color='red', linewidth=2))

plt.title('Box Plot: Global Active Power Distribution', fontsize=16, fontweight='bold')
plt.ylabel('Power (kW)', fontsize=12)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('boxplot_power.png', dpi=300, bbox_inches='tight')
plt.show()

print(" Graph 5 saved: boxplot_power.png")
print(" This shows the spread of data and identifies outliers (unusual high consumption)")
```

```python
correlation = df.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation, annot=True, fmt='.2f', cmap='coolwarm',
            square=True, linewidths=0.5, cbar_kws={"shrink": 0.8})
plt.title('Correlation Heatmap Between Electrical Features', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig('correlation_heatmap.png', dpi=300, bbox_inches='tight')
plt.show()

print(" Graph 6 saved: correlation_heatmap.png")
print(" This shows which electrical features are related to each other")
print("\n Strong correlations indicate features that change together")
```

```python
weekly_usage = df['Global_active_power'].resample('W').mean()

plt.figure(figsize=(14, 6))
plt.plot(weekly_usage, marker='o', linewidth=2, markersize=5, color='purple')
plt.title('Weekly Average Electricity Consumption', fontsize=16, fontweight='bold')
plt.xlabel('Week', fontsize=12)
plt.ylabel('Power (kW)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('weekly_consumption.png', dpi=300, bbox_inches='tight')
plt.show()

print(" Graph 7 saved: weekly_consumption.png")
print(" This shows week-by-week electricity usage trends")
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(df['Voltage'], bins=50, color='orange', edgecolor='black', alpha=0.7)
plt.title('Voltage Distribution', fontsize=14, fontweight='bold')
plt.xlabel('Voltage (V)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', alpha=0.3)

plt.subplot(1, 2, 2)
plt.plot(df['Voltage'].resample('D').mean(), linewidth=1, color='darkorange')
plt.title('Daily Average Voltage', fontsize=14, fontweight='bold')
plt.xlabel('Date', fontsize=12)
plt.ylabel('Voltage (V)', fontsize=12)
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('voltage_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

print(" Graph 8 saved: voltage_analysis.png")
print(" This analyzes voltage stability over time")
print(f" Average voltage: {df['Voltage'].mean():.2f} V")
```
